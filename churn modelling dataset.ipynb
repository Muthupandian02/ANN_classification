```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
data=pd.read_csv(r'D:\python_codes\deep_learning\Churn_Modelling.csv')
```

```python
df=data.copy()
```

```python
df.head()
```

```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
label_enc_gender=LabelEncoder()
onehot_enc_geo=OneHotEncoder()
df['Gender']=label_enc_gender.fit_transform(df['Gender'])
```

```python
geo_one_enc=onehot_enc_geo.fit_transform(df[['Geography']]).toarray()
geo_one_enc
```

```python
geo_enc_df=pd.DataFrame(geo_one_enc,columns=onehot_enc_geo.get_feature_names_out())
geo_enc_df
```

```python
df=pd.concat([df.drop('Geography',axis=1),geo_enc_df],axis=1)
```

```python
# saving the encoders
with open('label_enc_gender.pkl','wb') as file:
    pickle.dump(label_enc_gender,file)

with open('onehot_enc_geo','wb') as file:
    pickle.dump(onehot_enc_geo,file)
```

```python
x=df.drop('Exited',axis=1)
y=df['Exited']
```

```python
# dividing the dataset into dependent and independent
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=44)

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
with open('scaler','wb') as file:
    pickle.dump(scaler,file)
```

ANN implementation

when we initialize ANN neural network, the key parameters are,
      
      # sequential network
      
      # dense - it determines the no of  neurons (eg: 64, 64 no of neural nodes are created)
      
      # Activation function (sigmoid, tann, relu, leaky relu, parametric relu)
          
           - suggested, in hidden layer = relu and in output layer = sigmoid
      
      # optimizers -  useful in backward propagation for weight updatation
      
      # loss function (minimize)
      
      # metrics ----> accuracy, mse, mae
      
      # entire training information is stored -- in folders --and use tensor board for vizualization( displaying the logs )

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model=Sequential(
    [Dense(64,activation='relu',input_shape=((x_train.shape[1],))),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')]
)
```

```python
model.summary()
```

```python
model.compile (optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
log_dir='log/fit'
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```
